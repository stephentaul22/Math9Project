With this information, we can start to calculate things like average forest loss per year, forest loss per decade, etc. Also, if we consider outside information, like political institutional changes, we can make hypotheses about how these changes affected forest land cover in Brazil.

And here is an example of how to look at the forest loss in Brazil between 2000 and 2010.

```python
brazil = forest_area[forest_area['Entity'] == 'Brazil']
```

```python
brazil_2010 = float(brazil[brazil['Year'] == 2010]['Forest cover'])
```

```python
brazil_2000 = float(brazil[brazil['Year'] == 2000]['Forest cover'])
```

```python
diff = brazil_2000 - brazil_2010
```

```python
diff
```
